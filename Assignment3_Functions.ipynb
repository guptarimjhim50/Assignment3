**Question 1:** What is the difference between a function and a method in Python?

In Python, functions and methods are similar in that they both perform a block of code when called, but the key difference lies in how they are associated with objects.

**Function:**
A function is a block of code that is not associated with any object. You define it using the def keyword and call it independently.

**Method:**
A method is a function that is associated with an object. It is defined inside a class and is called on an instance (object) of that class. The first parameter of a method is usually self, which refers to the instance.

````python
def greet(name):
    return f"Hello, {name}!"

print(greet("ram"))
````

````python
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("ram"))
````

**Question 6:** Explain the concept of generators in Python and how they are defined?

A generator is a special type of **iterator** that allows you to generate **values on the fly**, instead of storing them all in memory at once.

**Generators are used when:**

1. You have a large dataset

2. You want to iterate one item at a time

3. You want to save memory and improve performance

4. Generators are defined using: Functions with yield instead of return

5. Generator expressions (like list comprehensions but with parentheses)

**Advantages of Generators:**

* Memory efficient: Doesn’t store the whole result in memory.
* Lazy evaluation: Values are produced only when needed.
* Great for large data processing, like reading big files or streams.

````python
def count_up_to(max):  #Functions with yield instead of return
    count = 1
    while count <= max:
        yield count
        count += 1

counter = count_up_to(3)

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
# print(next(counter))  # Raises StopIteration
````

````python
squares = (x * x for x in range(5))  #Generator expressions (like list comprehensions but with parentheses)
print(next(squares))  # Output: 0
print(next(squares))  # Output: 1
print(next(squares))  # Output: 4
print(next(squares))  # Output: 9
print(next(squares))  # Output: 16
````

**Question 7:** What are the advantages of using generators over regular functions?

1. **Memory Efficiency:**

   Regular function stores all results in memory (e.g., a full list).

   Generator produces one item at a time — no need to store the entire result in memory.

   Perfect for large datasets!

2. **Lazy Evaluation (On-Demand Values):**

   Generators only produce values when requested using next().

  This makes them faster to start and more responsive in real-time systems.

3.  **Cleaner Code for Iteration:**

 Generators simplify code that would otherwise need manual state tracking (e.g., using loops and intermediate lists).

4. **Infinite Sequences:**

 You can build infinite sequences with generators, which would be impossible (or crash your program) using a regular function.

5. **Pipeline Support:**

 Generators can be chained together for building data pipelines.

````python
def even_numbers(n):     #Cleaner Code for Iteration
    for i in range(n):
        if i % 2 == 0:
            yield i


def infinite_counter():     # Infinite Sequences:
    i = 0
    while True:
        yield i
        i += 1
````

**Question 8:** What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous function — basically a quick, one-line function without a name.

**A lambda function is defined using the lambda keyword:**

    lambda arguments: expression

it can have any number of arguments, but only one expression (no statements or multiple lines).

**Lambda functions are used when:**

**Question 10:** what is the difference between map(), reduce(), and filter() functions in Python?

map(), reduce(), and filter() are functional programming tools in Python. They all deal with processing iterables, but each has a different purpose.

**map()**:

* Transform Each Element

* Applies a function to every item in an iterable.

* Returns a new iterable with the results.

**Use it when:** You want to transform or change each item in a list.

**filter():**

* Keep Only Matching Items

* Applies a function that returns True/False.

* Keeps only items that return True.

**Use it when:** You want to transform or change each item in a list.

**reduce():**

* Combine Into a Single Value

* Applies a function to reduce the iterable to one value.

* Must be imported from functools.

**Use it when:** You want to accumulate a result (sum, product, etc.) from the list.

````python
from functools import reduce

nums = [1, 2, 3, 4, 5, 6]

# Step 1: Filter even numbers
filtered = filter(lambda x: x % 2 == 0, nums)

# Step 2: Square each
squared = map(lambda x: x ** 2, filtered)

# Step 3: Sum them all
result = reduce(lambda x, y: x + y, squared)

print(result)  # Output: 56 (2² + 4² + 6²)

#map
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, nums))
print(squares)  # Output: [1, 4, 9, 16]

#filter
nums = [1, 2, 3, 4]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # Output: [2, 4]


#reduce
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)  # Output: 24

````

**Question 4:** Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check up to √n
        if n % i == 0:
            return False
    return True
# is_prime(3)
````

**Question 5:** Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

````python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms  # Number of terms in the sequence
        self.current = 0  # The current term
        self.previous = 0  # The previous term
        self.next_term = 1  # The next term in the sequence
        self.count = 0  # Counter to track the number of terms generated

    def __iter__(self):
        return self  # This is the iterator itself

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration  # Stop iteration if the limit is reached
        else:
            # Return the next Fibonacci number
            if self.count == 0:
                self.count += 1
                return self.current
            elif self.count == 1:
                self.count += 1
                return self.next_term
            else:
                # Update Fibonacci terms
                fib = self.previous + self.next_term
                self.previous = self.next_term
                self.next_term = fib
                self.count += 1
                return fib
# obj = FibonacciIterator(10)
# for num in obj:
#     print(num)
````

````python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  #strip() to remove any trailing newlines or spaces
````

**Question 10:** Create a Python program that uses `filter()` to remove all the vowels from a given string.




````python
def remove_vowels(input_string):
    # Define a set of vowels
    vowels = "aeiouAEIOU"

    # Use filter to keep only non-vowel characters
    result = filter(lambda char: char not in vowels, input_string)

    # Join the filtered characters back into a string
    return ''.join(result)
````
